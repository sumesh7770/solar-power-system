## Data loading

### Subtask:
Load the dataset "solar_power_plant_modified (1).xlsx" into a pandas DataFrame.


**Reasoning**:
Load the excel file into a pandas DataFrame and display the first 5 rows.



```python
import pandas as pd

try:
    df = pd.read_excel('solar_power_plant_modified (1).xlsx')
    display(df.head())
except FileNotFoundError:
    print("Error: 'solar_power_plant_modified (1).xlsx' not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

## Data exploration

### Subtask:
Explore the loaded dataset to understand its characteristics.


**Reasoning**:
I need to explore the dataset's characteristics, including shape, data types, descriptive statistics, and missing values, as instructed.  This can be accomplished in a single code block.



```python
# Examine the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Get a summary of the DataFrame
print("\nDataFrame Info:")
df.info()

# Calculate descriptive statistics for numerical features
print("\nDescriptive Statistics for Numerical Features:")
display(df.describe())

# Analyze categorical features
print("\nValue Counts for Categorical Features:")
for col in df.select_dtypes(include=['object']).columns:
    print(f"\nColumn: {col}")
    display(df[col].value_counts())

# Check for missing values
print("\nMissing Values:")
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})
display(missing_df)
```

## Data analysis

### Subtask:
Perform a deeper analysis of the data, focusing on correlations and distributions.


**Reasoning**:
Calculate the correlation matrix, visualize it as a heatmap, and analyze the distribution of 'Production (kWh)' across different weather descriptions and time periods.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numerical features for correlation analysis
numerical_features = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numerical_features.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# Analyze the distribution of 'Production (kWh)' across different weather descriptions
weather_production = df.groupby('weatherDesc')['Production (kWh)'].mean()
plt.figure(figsize=(12, 6))
weather_production.plot(kind='bar', color='skyblue')
plt.title('Mean Production (kWh) by Weather Description')
plt.xlabel('Weather Description')
plt.ylabel('Mean Production (kWh)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Analyze the distribution of 'Production (kWh)' across different months
df['Month'] = df['Date'].dt.month
monthly_production = df.groupby('Month')['Production (kWh)'].mean()
plt.figure(figsize=(10, 5))
monthly_production.plot(kind='line', marker='o', color='green')
plt.title('Average Production (kWh) per Month')
plt.xlabel('Month')
plt.ylabel('Average Production (kWh)')
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
```

## Data visualization

### Subtask:
Visualize the data to gain further insights.


**Reasoning**:
Generate histograms, box plots, and scatter plots to visualize the data distribution and relationships between features.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms
plt.figure(figsize=(15, 10))
features = ['Production (kWh)', 'Temperature', 'Wind Speed', 'Humidity', 'Sunshine Hours']
for i, col in enumerate(features):
    plt.subplot(2, 3, i + 1)
    if col == 'Temperature':
        sns.histplot(df['tempC'], kde=True, color='skyblue')
    elif col == 'Wind Speed':
        sns.histplot(df['windSpeedKmph'], kde=True, color='orange')
    elif col == 'Humidity':
        sns.histplot(df['humidity'], kde=True, color='green')
    elif col == 'Sunshine Hours':
        sns.histplot(df['sunHour'], kde=True, color='red')
    elif col == 'Production (kWh)':
        sns.histplot(df['Production (kWh)'], kde=True, color='purple')
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()


# Box plots
plt.figure(figsize=(15, 5))
for i, col in enumerate(features):
    plt.subplot(1, 5, i + 1)
    if col == 'Temperature':
        sns.boxplot(y=df['tempC'], color='skyblue')
    elif col == 'Wind Speed':
        sns.boxplot(y=df['windSpeedKmph'], color='orange')
    elif col == 'Humidity':
        sns.boxplot(y=df['humidity'], color='green')
    elif col == 'Sunshine Hours':
        sns.boxplot(y=df['sunHour'], color='red')
    elif col == 'Production (kWh)':
        sns.boxplot(y=df['Production (kWh)'], color='purple')
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()


# Scatter plots
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.scatterplot(x='tempC', y='Production (kWh)', data=df, hue='Month')
plt.title('Production vs. Temperature')

plt.subplot(2, 2, 2)
sns.scatterplot(x='windSpeedKmph', y='Production (kWh)', data=df, hue='Month')
plt.title('Production vs. Wind Speed')

plt.subplot(2, 2, 3)
sns.scatterplot(x='sunHour', y='Production (kWh)', data=df, hue='Month')
plt.title('Production vs. Sunshine Hours')

plt.subplot(2, 2, 4)
sns.scatterplot(x='cloudcover', y='Production (kWh)', data=df, hue='Month')
plt.title('Production vs. Cloud Cover')

plt.tight_layout()
plt.show()

# Scatter plot matrix (pairplot)
sns.pairplot(df[['Production (kWh)', 'tempC', 'windSpeedKmph', 'sunHour', 'cloudcover']], diag_kind='kde')
plt.show()
```
